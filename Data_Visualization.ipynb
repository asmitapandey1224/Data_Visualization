```python
# Q3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 35, 30]

df = pd.DataFrame({'Category': categories, 'Value':values})

sns.barplot(x='Category', y='Value', data=df)

plt.title('Comparison of Categories Based on Numerical Values')
plt.xlabel('Category')
plt.ylabel('Value')

plt.show()
```

```python
# Q4.  Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

np.random.seed(42)
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [np.random.normal(loc=mean, scale=5, size=100) for mean in range(10,50,10)]

df = pd.DataFrame({
    'Category': np.repeat(categories, 100),
    'Value': np.concatenate(values)
})

sns.catplot(x='Category', y='Value', data=df, kind='box')
plt.title('Box Plot of Numerical Values Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()


sns.catplot(x='Category', y='Value', data=df, kind='violin')
plt.title('Violin Plot of Numerical Values Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()


sns.catplot(x='Category', y='Value', data=df, kind='boxen')
plt.title('boxen Plot of Numerical Values Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
# Q5.  Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

np.random.seed(42)
n_samples = 1000

x1 = np.random.normal(loc=0, scale=1, size=n_samples)
x2 = 2 * x1 + np.random.normal(loc=0, scale=1, size=n_samples)  # Strong positive correlation with x1
x3 = -1 * x1 + np.random.normal(loc=0, scale=1, size=n_samples) # Strong negative correlation with x1
x4 = np.random.normal(loc=0, scale=1, size=n_samples)

df = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

corr_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Q1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.

np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['X'], df['Y'], df['Z'], c='b', marker='o')

ax.set_title('3D Scatter Plot of Data Points')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
# 1. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

sales_pivot = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

fig = go.Figure(data=go.Heatmap(z=sales_pivot.values, x=sales_pivot.columns, y=sales_pivot.index, colorscale='YlGnBu'))

fig.update_layout(title='Heatmap of Sales Across Different Months and Days', xaxis=dict(title='Month'), yaxis=dict(title='Day'))

fig.show()
```

```python
# Q4. Using the given x and y data, generate a 3D surface plot to visualize the function z=sin(sqrt(x^2+y^2))

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title='3D Surface Plot of $z = \sin(\sqrt{x^2 + y^2})$',
                  scene=dict(
                      xaxis_title='X',
                      yaxis_title='Y',
                      zaxis_title='Z'
                  ))
fig.show()
```

```python
# Q5.  Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.

np.random.seed(25)
data = {'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],'Population': np.random.randint(100, 1000, 5), 'GDP': np.random.randint(500, 2000,5)}
df = pd.DataFrame(data)

fig = go.Figure()

for i, row in df.iterrows():
    fig.add_trace(
        go.Scatter(
            x=[row['GDP']],
            y=[row['Population']],
            mode='markers',
            marker=dict(size=row['Population'], sizemode='area'),
            name=row['Country']
        )
    )

fig.update_layout(
    title='Bubble Chart of GDP vs Population for Each Country',
    xaxis=dict(title='GDP'),
    yaxis=dict(title='Population'),
)

fig.show()
```

```python
# Q1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title='Sine Wave', x_axis_label='x', y_axis_label='sin(x)', width=600, height=400)
p.line(x, y, line_width=2, color='navy')

show(p)
```

```python
# Q2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

from bokeh.plotting import figure, show
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.transform import factor_cmap

np.random.seed(0)
n = 100
x = np.random.rand(n) * 10
y = np.random.rand(n) * 10
sizes = np.random.rand(n) * 10 + 5
colors = np.random.randint(0, 256, n)

p = figure(title='Scatter Plot', x_axis_label='X', y_axis_label='Y', width=600, height=400)

p.scatter(x, y, size=sizes, fill_color='red', fill_alpha=0.6)

show(p)
```

```python
# Q3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset
from bokeh.models import ColumnDataSource

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts, color=['blue', 'orange', 'green', 'yellow']))

p = figure(x_range=fruits, height=350, title="Fruit Counts", toolbar_location=None, tools="")

p.vbar(x='fruits', top='counts', width=0.5, color='color', legend_field='fruits', source=source)

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Counts"
p.xaxis.axis_label = "Fruits"
p.xaxis.major_label_orientation = 1.2

show(p)
```

```python
# Q4. Create a Bokeh histogram to visualize the distribution of the given data.

from bokeh.transform import dodge

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram", x_axis_label="Value", y_axis_label="Frequency", width=600, height=400)

p.vbar(x=dodge('edges', 0, range=p.x_range), top='hist', width=0.1, color="navy", legend_label="Histogram", source=dict(hist=hist, edges=edges))

p.legend.location = "top_right"
p.legend.click_policy="hide"

show(p)
```

```python
# Q5. Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

p = figure(title="Heatmap", x_axis_label="X", y_axis_label="Y", width=600, height=400)

colors = ['#440154', '#3B528B', '#21918C', '#5EC962', '#FDE725']
mapper = linear_cmap(field_name='data_heatmap', palette=colors, low=data_heatmap.min(), high=data_heatmap.max())

p.rect(x='xx', y='yy', width=1, height=1, source={'xx': xx.flatten(), 'yy': yy.flatten(), 'data_heatmap': data_heatmap.flatten()}, fill_color=mapper)

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
```
